```python
imagen = cv2.imread('cubo2.jpg', cv2.IMREAD_UNCHANGED)
original = cv2.imread("cubo2.jpg")
```

```python
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
cv2_imshow(gris)
```

```python

# Aplicar suavizado Gaussiano
gauss = cv2.GaussianBlur(gris, (5,5), 0)
 
cv2_imshow(gauss)
```

```python
canny = cv2.Canny(gauss, 50, 150)
 
cv2_imshow(canny)
 
```

```python
rojo_abajo = (0,150,50)
rojo_arriba = (10,255,250)
maskrojo = cv2.inRange(img_hsv, rojo_abajo, rojo_arriba)
imarojo = cv2.bitwise_and(img_hsv, img_hsv, mask=maskrojo)
(contornos,_) = cv2.findContours(maskrojo, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(original, contornos, -1, (255,0,0), 3)
for c in contornos:
      area = cv2.contourArea(c)
      if area > 3000:
        M = cv2.moments(c)
        if (M["m00"]==0): M["m00"]=1
        x = int(M["m10"]/M["m00"])
        y = int(M['m01']/M['m00'])
        cv2.circle(original, (x,y), 7, (0,255,0), -1)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(original,'rojo',(x-10,y-10),font, 0.75,(0,255,0),1,cv2.LINE_AA)
   
plt.subplot(1,2,1)
plt.imshow(maskrojo, cmap= 'gray')
plt.subplot(1,2,2)
plt.imshow(imarojo)
```
